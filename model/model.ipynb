```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import datasets, transforms
import random
from tqdm import tqdm
import matplotlib.pyplot as plt

# Define the encoders
class Encoder(nn.Module):
    def __init__(self):
        super(Encoder, self).__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(True),
            nn.Conv2d(64, 128, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(True),
            nn.Conv2d(128, 256, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(True)
        )
        self.fc = nn.Linear(256*4*4, 512)

    def forward(self, x):
        x = self.conv(x)
        x = x.view(x.size(0), -1)
        x = self.fc(x)
        return x

# Define the decoders for CIFAR-10 and MNIST
class CIFARDecoder(nn.Module):
    def __init__(self):
        super(CIFARDecoder, self).__init__()
        self.fc = nn.Linear(512, 256*4*4)
        self.deconv = nn.Sequential(
            nn.ConvTranspose2d(256, 128, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(True),
            nn.ConvTranspose2d(128, 64, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(True),
            nn.ConvTranspose2d(64, 3, kernel_size=4, stride=2, padding=1),
            nn.Tanh()
        )

    def forward(self, x):
        x = self.fc(x)
        x = x.view(x.size(0), 256, 4, 4)
        x = self.deconv(x)
        return x

class MNISTDecoder(nn.Module):
    def __init__(self):
        super(MNISTDecoder, self).__init__()
        self.fc = nn.Linear(512, 256*4*4)
        self.deconv = nn.Sequential(
            nn.ConvTranspose2d(256, 128, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(True),
            nn.ConvTranspose2d(128, 64, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(True),
            nn.ConvTranspose2d(64, 3, kernel_size=4, stride=2, padding=1),
            nn.Tanh()
        )

    def forward(self, x):
        x = self.fc(x)
        x = x.view(x.size(0), 256, 4, 4)
        x = self.deconv(x)
        return x

class MultiModalNetwork(nn.Module):
    def __init__(self):
        super(MultiModalNetwork, self).__init__()
        self.encoder = Encoder()
        self.cifar_decoder = CIFARDecoder()
        self.mnist_decoder = MNISTDecoder()

    def forward(self, combined_img):
        features = self.encoder(combined_img)
        cifar_img = self.cifar_decoder(features)
        mnist_img = self.mnist_decoder(features)
        return cifar_img, mnist_img

# Custom Dataset
class CombinedDataset(Dataset):
    def __init__(self, cifar_data, mnist_data, num_samples=300):
        self.cifar_data = cifar_data
        self.mnist_data = mnist_data
        self.num_samples = num_samples
        
        # Randomly select num_samples from each dataset
        self.cifar_indices = random.sample(range(len(self.cifar_data)), self.num_samples)
        self.mnist_indices = random.sample(range(len(self.mnist_data)), self.num_samples)
        
    def __len__(self):
        return self.num_samples

    def __getitem__(self, idx):
        # Get CIFAR-10 image and label
        cifar_img, _ = self.cifar_data[self.cifar_indices[idx]]
        # Get MNIST image and label
        mnist_img, _ = self.mnist_data[self.mnist_indices[idx]]

        # Convert MNIST image to 3 channels
        mnist_img = mnist_img.repeat(3, 1, 1)
        
        # Sum CIFAR-10 and MNIST images and divide by 2 to get the mean
        combined_img = (cifar_img + mnist_img) / 2

        return combined_img, cifar_img, mnist_img

# Transformations for CIFAR-10 and MNIST
transform_cifar = transforms.Compose([
    transforms.ToTensor()
])

transform_mnist = transforms.Compose([
    transforms.ToTensor(),
    transforms.Resize((32, 32))  # Resize MNIST images to 32x32
])

# Load CIFAR-10 and MNIST datasets
cifar_dataset = datasets.CIFAR10(root='./data', train=True, download=True, transform=transform_cifar)
mnist_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform_mnist)

# Create CombinedDataset
combined_dataset = CombinedDataset(cifar_dataset, mnist_dataset, num_samples=300)

# Hyperparameters
batch_size = 64
lr = 0.0002
num_epochs = 200

# DataLoader
dataloader = DataLoader(combined_dataset, batch_size=batch_size, shuffle=True)

# Initialize the network and optimizer
multi_modal_net = MultiModalNetwork().cuda()
optimizer = optim.Adam(multi_modal_net.parameters(), lr=lr)
criterion_mse = nn.MSELoss()

# Function to plot images
def plot_images(epoch, combined_img, reconstructed_cifar_img, reconstructed_mnist_img):
    fig, axes = plt.subplots(3, 2, figsize=(8, 12))
    for i in range(2):
        axes[0, i].imshow(combined_img[i].permute(1, 2, 0).cpu().detach().numpy())
        axes[0, i].set_title(f'Combined Image {i+1}')
        axes[1, i].imshow(reconstructed_cifar_img[i].permute(1, 2, 0).cpu().detach().numpy())
        axes[1, i].set_title(f'Reconstructed CIFAR Image {i+1}')
        axes[2, i].imshow(reconstructed_mnist_img[i].permute(1, 2, 0).cpu().detach().numpy())
        axes[2, i].set_title(f'Reconstructed MNIST Image {i+1}')
    plt.tight_layout()
    plt.show()

# Training loop
for epoch in range(num_epochs):
    epoch_loss = 0
    for combined_img, cifar_img, mnist_img in tqdm(dataloader, desc=f"Epoch {epoch+1}/{num_epochs}"):
        combined_img, cifar_img, mnist_img = combined_img.cuda(), cifar_img.cuda(), mnist_img.cuda()
        
        # Train the network
        optimizer.zero_grad()

        reconstructed_cifar_img, reconstructed_mnist_img = multi_modal_net(combined_img)
        
        loss_reconstruction_cifar = criterion_mse(reconstructed_cifar_img, cifar_img)
        loss_reconstruction_mnist = criterion_mse(reconstructed_mnist_img, mnist_img)

        loss = loss_reconstruction_cifar + loss_reconstruction_mnist
        loss.backward()
        optimizer.step()

        epoch_loss += loss.item()

    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss/len(dataloader):.4f}")

    # Plot images after each epoch
    combined_img, cifar_img, mnist_img = next(iter(dataloader))
    combined_img, cifar_img, mnist_img = combined_img.cuda(), cifar_img.cuda(), mnist_img.cuda()
    reconstructed_cifar_img, reconstructed_mnist_img = multi_modal_net(combined_img)
    if epoch % 20 == 0:
        plot_images(epoch, combined_img, reconstructed_cifar_img, reconstructed_mnist_img)
```
